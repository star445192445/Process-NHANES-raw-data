```python
#血压
### SEQN    编码
### BPQ020  自我报告高血压 1有； 2没有； NAN缺失
### BPXSY1  BPXSY2  BPXSY3  三次收缩压
### BPXDI1  BPXDI2  BPXDI3  三次舒张压
### systolic blood	   收缩压平均值
### diastoloc blood	   舒张压平均值
### hypertension_level  是否有高血压 1有 0没有 NA 缺失

# 身体测量
### BMXBMI    BMI

# 脂蛋白
### LBDHDDSI	    高密度脂蛋白
### LBDTCSI       总胆固醇
### non_HDL       非高密度脂蛋白

# CRP
### LBXCRP    C反应蛋白(mg/dl)

# 基本口人学
### 'RIDAGEYR'     年龄
### 'RIDRETH1'     种族
### 'DMDEDUC2'     教育水平
### 'INDFMPIR'     家庭年收入与贫困线的比值

# 糖尿病
### DIQ010          自我报告是否有糖尿病
### LBDGLUSI        空腹血糖
### LBDGLTSI        oggt2小时血糖
### diabetes_level  糖尿病等级  0没有 1有 2糖尿病前期

#饮食
### diet score  饮食评分

# 近亲
### MCQ300A      近亲心脏病
### MCQ300C      近亲糖尿病

# mental
### mental level    精神状况评级 0-3

# 死亡数据
### eligstat        是否链接上  1.是    2.18岁以下
### mortstat        生存状态    1.存活  2.死亡
### ucod_leading    死亡原因    1.心脏病；2恶性肿瘤；3.慢性呼吸系统疾病；4.意外；5.脑血管疾病；6.阿尔兹海默症；7.糖尿病；8.流感和肺炎；9.肾病；10.其他原因
### permth_int      从访谈到死亡的月数
### permth_exm      从MEC到死亡的月数
### follow_year     #随访年数

# 运动
###physical_level    运动等级
###Sedentary_minutes   静坐时间-mins

# 睡眠
### SLD010H 睡眠时长
### sleep_level 睡眠时长评级  0不好；1一般；2良好
```

```python
features = [
    'SEQN',
    'hypertension_level',   #是否有高血压 0没有 1有  NA 缺失
    'BMXBMI',   #BMI
    'non_HDL',  #非高密度脂蛋白
    'LBXCRP',   #CRP
    'RIDAGEYR', #年龄
    'RIAGENDR', #性别
    'RIDRETH1', #种族
    'DMDEDUC2', #教育水平
    'INDFMPIR', #收入与贫困线的比值
    'diabetes_level',   #是否有糖尿病   0没有 1有 2糖尿病前期
    'diet_score',   #饮食评分
    'MCQ300A',  #近亲心脏病
    'MCQ300C',  #近亲糖尿病
    'mental_level', #精神状态0-3
    'mortstat', #生存状态
    'ucod_leading',  #死亡原因    1.心脏病；2恶性肿瘤；3.慢性呼吸系统疾病；4.意外；5.脑血管疾病；6.阿尔兹海默症；7.糖尿病；8.流感和肺炎；9.肾病；10.其他原因
    'follow_year',   #随访年数
    'physical_level',
    'Sedentary_minutes',   
    'sleep_level',  #睡眠时长评级  0不好；1一般；2良好
    'drinks',
    'smoking_status',
    'equal minutes',

    'LBXGH',        #糖化血红蛋白
    'MCQ220',       #癌症
    'MCQ160C',      #冠心病
    'MCQ160F',      #中风
    
]
```

```python
import pandas as pd
```

```python
# 血压数据
hypertension = pd.read_excel('E:/work/Paper/NHANES/0910Analysis Data/Final data/blood_pressure.xlsx', index_col=0)
# BMI
body_measure = pd.read_excel('E:/work/Paper/NHANES/0910Analysis Data/Final data/body_measure.xlsx', index_col=0)
# 非高密度脂蛋白
chelesterol = pd.read_excel('E:/work/Paper/NHANES/0910Analysis Data/Final data/chelesterol.xlsx', index_col=0)
# CRP
crp = pd.read_excel('E:/work/Paper/NHANES/0910Analysis Data/Final data/crp.xlsx', index_col=0)
# demo
demographic = pd.read_excel('E:/work/Paper/NHANES/0910Analysis Data/Final data/demographic.xlsx', index_col=0)
# 糖尿病
diabetes = pd.read_excel('E:/work/Paper/NHANES/0910Analysis Data/Final data/diabetes.xlsx', index_col=0)
# 饮食
diet = pd.read_excel('E:/work/Paper/NHANES/0910Analysis Data/Final data/DGA_HEI_score.xlsx', index_col=0)
# 疾病史
medical_conditions = pd.read_excel('E:/work/Paper/NHANES/0910Analysis Data/Final data/medical_conditions.xlsx', index_col=0)
# 精神
mental_health_evaluate = pd.read_excel('E:/work/Paper/NHANES/0910Analysis Data/Final data/mental_health_level.xlsx', index_col=0) 
# 死亡数据
mortality = pd.read_excel('E:/work/Paper/NHANES/0910Analysis Data/Final data/mortality.xlsx', index_col=0)
# 体力活动
physical = pd.read_excel('E:/work/Paper/NHANES/0910Analysis Data/Final data/physical_level.xlsx', index_col=0)
# 睡眠
sleep = pd.read_excel('E:/work/Paper/NHANES/0910Analysis Data/Final data/sleep_level.xlsx', index_col=0)
# 饮酒
alcohol = pd.read_excel('E:/work/Paper/NHANES/0910Analysis Data/Final data/alcohol_use.xlsx', index_col=0)
# 吸烟
smoking = pd.read_excel('E:/work/Paper/NHANES/0910Analysis Data/Final data/smoking_status.xlsx', index_col=0)
# 糖化血红蛋白
hba1c = pd.read_excel('E:/work/Paper/NHANES/0910Analysis Data/Final data/glycohemoglobin.xlsx', index_col=0)
```

```python
merge_list = [
    hypertension, 
    body_measure, 
    chelesterol, 
    crp, 
    demographic, 
    diabetes,
    diet, 
    medical_conditions, 
    mental_health_evaluate, 
    physical,
    sleep,
    alcohol,
    smoking,
    hba1c
  ]

data = mortality
for i in merge_list:
    data = pd.merge(left=data, right=i, on='SEQN', how='outer')
```

```python
data_0910 = data[features]
```

```python
data_0910 = data_0910[pd.notna(data_0910['mortstat'])]
```
