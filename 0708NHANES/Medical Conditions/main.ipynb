```python
### MCQ300A      #近亲心脏病
### MCQ300C      #近亲糖尿病
### MCQ220       #是否有被告知有癌症
### MCQ160C      #是否有冠心病   #1有 2没有
### MCQ160F      #中风          #1有 2没有

import pandas as pd
import numpy as np
```

```python
medical_conditions_df = pd.read_sas('E:/work/Paper/NHANES/0708Analysis Data/Medical Conditions/MCQ_E.XPT')
```

```python
medical_conditions_extract_features = [
    'SEQN',
    'MCQ300A',      #近亲心脏病
    'MCQ300C',      #近亲糖尿病
    'MCQ220',       #癌症
    'MCQ160C',      #冠心病
    'MCQ160F',      #中风
]
medical_conditions_extract_df = medical_conditions_df[medical_conditions_extract_features]
medical_conditions_extract_df['MCQ300A'].replace([7,9], np.nan, inplace=True)
medical_conditions_extract_df['MCQ300C'].replace([7,9], np.nan, inplace=True)
medical_conditions_extract_df['MCQ220'].replace([7,9, np.nan], np.nan, inplace=True)
medical_conditions_extract_df['MCQ160C'].replace([7,9, np.nan], 2, inplace=True)
medical_conditions_extract_df['MCQ160F'].replace([7,9, np.nan], 2, inplace=True)
```
