```python
# LBDHDDSI	    高密度脂蛋白
# LBDTCSI       总胆固醇
# LBDTRSI       甘油三酯
# non_HDL       非高密度脂蛋白

import pandas as pd
import numpy as np
```

```python
# 甘油三酯
tri_df = pd.read_sas('E:/work/Paper/NHANES/0708Analysis Data/Cholesterol/TRIGLY_E.XPT')
tri_extract_features = [
    'SEQN',
    'LBDTRSI'
]
tri_extract_df = tri_df[tri_extract_features]
tri_extract_df
```

```python
# 高密度脂蛋白
hdl_df = pd.read_sas('E:/work/Paper/NHANES/0708Analysis Data/Cholesterol/HDL_E.XPT')
hdl_extract_features = [
    'SEQN',
    'LBDHDDSI',     #直接测量的高密度脂蛋白(mmol/L)
]
hdl_extract_df = hdl_df[hdl_extract_features]
```

```python
# 总胆固醇
total_cholesterol_df = pd.read_sas('E:/work/Paper/NHANES/0708Analysis Data/Cholesterol/TCHOL_E.XPT')

total_cholesterol_extract_features = [
    'SEQN',
    'LBDTCSI',      #胆固醇总量( mmol/L)
]
total_cholesterol_extract_df = total_cholesterol_df[total_cholesterol_extract_features]
```

```python
# 合并文件
chelesterol_df = pd.merge(left=hdl_extract_df, right=total_cholesterol_extract_df, on='SEQN', how='outer')
chelesterol_df = pd.merge(left=chelesterol_df, right=tri_extract_df, on='SEQN', how='outer')
chelesterol_df['non_HDL'] = chelesterol_df['LBDTCSI'] - chelesterol_df['LBDHDDSI']
chelesterol_df
```

```python
chelesterol_df.to_excel('E:/work/Paper/NHANES/0708Analysis Data/Final data/chelesterol.xlsx')
```
